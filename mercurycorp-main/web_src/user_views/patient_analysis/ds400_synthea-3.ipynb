# Data PreProcessing

## Joining Tables

```python
# loading the CSVs
patients = dataframes['patients.csv']
encounters = dataframes['encounters.csv']
conditions = dataframes['conditions.csv']
procedures = dataframes['procedures.csv']
medications = dataframes['medications.csv']
claims = dataframes['claims.csv']
observations = dataframes['observations.csv']

# rename columns to avoid conflicts before merging
patients = patients.rename(columns=lambda x: f"PATIENT_{x}" if x != "Id" else "PATIENT_ID")
encounters = encounters.rename(columns=lambda x: f"ENCOUNTER_{x}" if x != "Id" else "ENCOUNTER_ID")
conditions = conditions.rename(columns=lambda x: f"CONDITION_{x}" if x != "ENCOUNTER" else "ENCOUNTER_ID")
procedures = procedures.rename(columns=lambda x: f"PROCEDURE_{x}" if x != "ENCOUNTER" else "ENCOUNTER_ID")
medications = medications.rename(columns=lambda x: f"MEDICATION_{x}" if x != "ENCOUNTER" else "ENCOUNTER_ID")
claims = claims.rename(columns=lambda x: f"CLAIM_{x}" if x != "PATIENTID" else "PATIENT_ID")
observations = observations.rename(columns=lambda x: f"OBSERVATION_{x}" if x != "ENCOUNTER" else "ENCOUNTER_ID")

# merge Encounters with Patients
df = encounters.merge(patients, left_on='ENCOUNTER_PATIENT', right_on='PATIENT_ID', how='left')

# merge Conditions
df = df.merge(conditions, on='ENCOUNTER_ID', how='left')

# merge Procedures
df = df.merge(procedures, on='ENCOUNTER_ID', how='left')

# merge Medications
df = df.merge(medications, on='ENCOUNTER_ID', how='left')

# merge Observations
df = df.merge(observations, on='ENCOUNTER_ID', how='left')

# drop redundant ID cols
df.drop(columns=['PATIENT_ID', 'ENCOUNTER_ID'], errors='ignore', inplace=True)

print("Shape:", df.shape)
df.head()
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical

# USED AI AS TEMPLATE

# Step 1: Sort encounters by patient & time
df_filtered = df_filtered.sort_values(by=["PATIENT_SSN", "ENCOUNTER_START"])

# Step 2: Convert `ENCOUNTER_DESCRIPTION` into numerical values
encounter_to_index = {enc: i for i, enc in enumerate(df_filtered["ENCOUNTER_DESCRIPTION"].unique())}
df_filtered["ENCOUNTER_DESCRIPTION_IDX"] = df_filtered["ENCOUNTER_DESCRIPTION"].map(encounter_to_index)

# Step 3: Create sequences
X_seq = []  # Input sequences
y_seq = []  # Target (Next encounter)

grouped = df_filtered.groupby('PATIENT_SSN')

for _, group in grouped:
    encounter_ids = group["ENCOUNTER_DESCRIPTION_IDX"].values  # Convert encounters to IDs

    for i in range(1, len(encounter_ids)):
        X_seq.append(encounter_ids[:i])  # Past encounters
        y_seq.append(encounter_ids[i])   # Next encounter

# Step 4: Pad sequences to ensure uniform length
X_seq = pad_sequences(X_seq, padding='pre')  # Padding at the beginning
y_seq = np.array(y_seq)

# Step 5: Convert target variable into categorical
y_seq = to_categorical(y_seq, num_classes=len(encounter_to_index))

# Print dataset shape
print(f"Shape of X_seq: {X_seq.shape}")  # (num_sequences, max_seq_length)
print(f"Shape of y_seq: {y_seq.shape}")  # (num_sequences, num_classes)

# Save mapping for later decoding
index_to_encounter = {v: k for k, v in encounter_to_index.items()}
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding

# LSTM MODEL
model = Sequential([
    Embedding(input_dim=len(encounter_to_index), output_dim=64, input_length=X_seq.shape[1]),  # Embedding layer
    LSTM(128, return_sequences=False),  # LSTM layer
    Dense(len(encounter_to_index), activation="softmax")  # Output layer
])

# Compile
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

# Train
model.fit(X_seq, y_seq, epochs=10, batch_size=32, validation_split=0.2)

# Save
model.save("lstm_encounter_model.h5")
```

```python
train_loss, train_acc = model.evaluate(X_seq, y_seq, verbose=1)

print(f"Model Training Accuracy: {train_acc:.4f}")
print(f"Model Training Loss: {train_loss:.4f}")
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical

df_filtered = df_filtered.sort_values(by=["PATIENT_SSN", "ENCOUNTER_START"])

encounter_to_index = {enc: i for i, enc in enumerate(df_filtered["ENCOUNTER_DESCRIPTION"].unique())}
df_filtered["ENCOUNTER_DESCRIPTION_IDX"] = df_filtered["ENCOUNTER_DESCRIPTION"].map(encounter_to_index)

X_seq = []
y_seq = []

grouped = df_filtered.groupby('PATIENT_SSN')

for _, group in grouped:
    encounter_ids = group["ENCOUNTER_DESCRIPTION_IDX"].values

    for i in range(1, len(encounter_ids)):
        X_seq.append(encounter_ids[:i])
        y_seq.append(encounter_ids[i])

X_seq = pad_sequences(X_seq, padding='pre')
y_seq = np.array(y_seq)

y_seq = to_categorical(y_seq, num_classes=len(encounter_to_index))

index_to_encounter = {v: k for k, v in encounter_to_index.items()}

print(f"Shape of X_seq: {X_seq.shape}")  # (num_sequences, max_seq_length)
print(f"Shape of y_seq: {y_seq.shape}")  # (num_sequences, num_classes)
```

```python
import tensorflow as tf
import tensorflow.keras.backend as K

# Define focal loss function manually
def focal_loss(alpha=0.25, gamma=2.0):
    def loss(y_true, y_pred):
        y_pred = K.clip(y_pred, 1e-7, 1 - 1e-7)  # Prevent log(0) error
        cross_entropy = -y_true * K.log(y_pred)
        weight = alpha * K.pow((1 - y_pred), gamma)
        focal_loss = weight * cross_entropy
        return K.sum(focal_loss, axis=1)  # Sum across all classes
    return loss
```

```python
def create_sequences(df):
    X_seq = []
    y_seq = []

    grouped = df.groupby('PATIENT_SSN')

    for _, group in grouped:
        encounter_ids = group["ENCOUNTER_DESCRIPTION_IDX"].values

        for i in range(1, len(encounter_ids)):
            X_seq.append(encounter_ids[:i])
            y_seq.append(encounter_ids[i])

    # Pad sequences
    X_seq = pad_sequences(X_seq, padding='pre')
    y_seq = np.array(y_seq)

    # One-hot encode target variable
    y_seq = to_categorical(y_seq, num_classes=len(encounter_to_index))

    return X_seq, y_seq
```

```python
### Splitting Data by Patient
from sklearn.model_selection import train_test_split

unique_patients = df_filtered["PATIENT_SSN"].unique()

# Split 80/20
train_patients, val_patients = train_test_split(unique_patients, test_size=0.2, random_state=101)

df_train = df_filtered[df_filtered["PATIENT_SSN"].isin(train_patients)]
df_val = df_filtered[df_filtered["PATIENT_SSN"].isin(val_patients)]

# Convert to sequences
X_train, y_train = create_sequences(df_train)
X_val, y_val = create_sequences(df_val)
```

```python
print(f"Training Set: {X_train.shape}, {y_train.shape}")
print(f"Validation Set: {X_val.shape}, {y_val.shape}")
```

```python
### Modeling
import keras_tuner as kt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard

# Function to build the LSTM model with tunable hyperparameters
def build_lstm_model_focal_loss(hp):
    model = Sequential()
    model.add(Embedding(input_dim=len(encounter_to_index),
                        output_dim=hp.Int('embedding_dim', 32, 128, step=32),
                        input_length=X_seq.shape[1]))
    model.add(LSTM(hp.Int('lstm_units', 64, 256, step=64), return_sequences=False))
    model.add(Dropout(hp.Float('dropout_rate', 0.2, 0.5, step=0.1)))
    model.add(Dense(len(encounter_to_index), activation="softmax"))

    model.compile(
        optimizer=Adam(learning_rate=hp.Float('learning_rate', 1e-4, 1e-2, sampling='log')),
        loss=focal_loss(alpha=0.25, gamma=2.0),  # Using our defined function
        metrics=["accuracy"]
    )

    return model


tuner = kt.RandomSearch(
    build_lstm_model_focal_loss,
    objective="val_accuracy",
    max_trials=10,
    executions_per_trial=1,
    directory="lstm_tuning_focal_loss",
    project_name="lstm_focal_loss_tuning"
)

# Early stopping for overfitting
early_stopping = EarlyStopping(monitor="val_loss", patience=3, restore_best_weights=True)

# Train with Focal Loss
tuner.search(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_val, y_val), callbacks=[early_stopping])

# Get the best hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print(f"Best Hyperparameters: {best_hps.values}")

# Train the best model
best_lstm_model_focal = tuner.hypermodel.build(best_hps)
best_lstm_model_focal.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_val, y_val), callbacks=[early_stopping])

# Save the focal loss model
best_lstm_model_focal.save("lstm_encounter_model_focal.h5")
```
